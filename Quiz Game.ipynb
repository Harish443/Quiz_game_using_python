```python

def new_game():
    guesses = []
    correct_guess = 0
    question_num = 1
    
    print("HERE'S YOUR QUESTIONS")
    for quest in questions:
        print("---------------------------------")
        print(quest)
        for opt in options[question_num - 1]:
            print(opt)
        print()
        guess = input("Enter your option : ").upper()
        if guess not in ["A","B","C","D"]:
            print("Invalid Option")
        guesses.append(guess)
        
        correct_guess += check_answer(questions.get(quest),guess)
        question_num += 1
        
    display_score(correct_guess,guesses)
        
def check_answer(answer,guess):
    
    if answer == guess:
        print("Correct")
        print()
        return 1
    else:
        print("Wrong")
        print()
        return 0

def display_score(correct_guess,guesses):
    print("---------------------------------")
    print("RESULTS :")
    print("---------------------------------")
    
    print("Answers : ",end ="")
    for i in questions:
        print(questions.get(i),end = " ")
    print()
    
    print("Guesses : ",end ="")
    for i in guesses:
        print(i,end = " ")
    print()

    score = int((correct_guess/len(questions))*100)
    print("your Score is: ",str(score),"%")
    print()
    
def play_again():
    
    response = input("Do you want to play again :").upper()
    if response == "YES":
        return True
    else:
        return False

questions = {"Who created python ? " : "A",
             "what is the abrevation of AI ? " : "C",
             "Will Ai replace evey job ? " : "B",
             "which year python was invented ? " : "D",
             "what is the abrevation of DC ? ": "A"}

options =[["A. Guido Van Rossum" , "B. Hari Prasad" , "C. Ramanajum" , "D. Kepler"],
          ["A. Auto Increment" , "B. Auto Intellectual" , "C. Artificial Intelligence" , "D. Automated Intelligent"],
          ["A. yes" , "B. no" , "C. maybe" , "D. confirm"],
          ["A. 1980" , "B. 1982" , "C. 1990" , "D. 1991"],
          ["A. Distributed Computing" , "B. Data connection" , "C. Data Creativity" , "D. Distributed Components"]] 
new_game()

while play_again():
    new_game()
    
print("BYEEEEEE")
```
